```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in the data
FileToRead = 'BikeFahrrad_GoogleShopping_Bayern.csv'
df = pd.read_csv(FileToRead)

# Rename Column 'Mes' to 'Month'
df.rename(columns={'Mes':'Month'}, inplace=True)

# Change the Month column to a datetime object
df['Month'] = pd.to_datetime(df['Month'])

# Only keep Year and Month from the datetime object
df['Month'] = df['Month'].dt.to_period('M')

# Print the first 5 rows of the dataframe
print(df.head())
```

```python
# Create a new data frame with only the columns 'Month' and 'Total' by adding the values of the columns 'Bike' and 'Fahrrad'
df2 = df.copy()
df2['Total'] = df2['Bike'] + df2['fahrrad']

# drop the columns 'Bike' and 'fahrrad'
data = df2.drop(columns=['Bike', 'fahrrad'])

# Print the first 5 rows of the new dataframe
print(data.head())
data = data.set_index('Month')
```

```python
# Plot the data
# set plot size and title of the plot 
data.plot(figsize=(10, 5), title='Bike and Fahrrad Searches in the Google Shopping section - in Bayern')
```
